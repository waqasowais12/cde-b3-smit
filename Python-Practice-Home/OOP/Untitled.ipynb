### Inheritance

```python
class Parent():
    def __init__(self, eye_color, hair_color):
        self.eye_color = eye_color
        self.hair_color = hair_color
    
    def can_paint(self):
        print("Have ability to draw/paint art")
```

```python
par_obj = Parent("Black", "Brown")
print(par_obj.eye_color)
print(par_obj.hair_color)
par_obj.can_paint()
```

```python
class Child(Parent):
    def __init__(self, eye_color, hair_color, skin_color):
        super().__init__(eye_color, hair_color)
        self.skin_color = skin_color
    
    def can_drive(self):
        print("Have ability to drive")
```

```python
child_obj = Child("Black", "Black", "Brown")
child_obj.can_paint()
child_obj.can_drive()
print(child_obj.eye_color)
print(child_obj.hair_color)
print(child_obj.skin_color)
```

**Task**

- Create a class with name "Father" which have two attributes and one method.
- Create a class with name "Mother" which have one attribute.
- Create a class with name "Child" that inherit characteristics of "Father" and "Mother" but should have atleast two more attributes and methods other than "Father class". 

```python
class Father():
    def __init__(self, name, cast):
        self.name = name
        self.cast = cast
    
    def hardWorker(self):
        print("Have ability to work hard.")
```

```python
class Mother():
    def __init__(self, smile):
        self.smile = smile
```

```python
class Child(Father, Mother):
    def __init__(self, name, cast, smile, body, age):
        Father.__init__(self, name, cast)
        Mother.__init__(self, smile)
        self.boby = body
        self.age = age
    
    def can_play(self):
        print("Have ability to play cricket")
```

```python
child_objj = Child("Waqas", "Khan", "Beautifull", "Normal", "23")
child_objj.hardWorker()
child_objj.can_play()
print(child_objj.name)
print(child_objj.cast)
print(child_objj.age)
print(child_objj.boby)
print(child_objj.smile)
```

### Overriding

```python
class Animal():
    def eating(self):
        print("Animal can eat.")

class Birds(Animal):
    def eating(self): #method overriding
        print("Birds can eat.")
```

```python
ani_obj = Animal()
ani_obj.eating()
```

```python
birds_obj = Birds()
birds_obj.eating()
```

### Overloading

```python
from typing import overload

class Adder():

    @overload
    def add(self, x:int, y:int) -> int:
        ...
    @overload
    def add(self, x:float, y:float) -> float:
        ...
    @overload
    def add(self, x:str, y:str) -> str:
        ...

    def add(self, x, y):
        if isinstance(x, int) and isinstance(y, int):
            return x+y
        elif isinstance(x, float) and isinstance(y, float):
            return x+y
        elif isinstance(x, str) and isinstance(y, str):
            return x+y
        else:
            print("Type Error!")
```

```python
add_obj = Adder()
add_obj.add(2,3)
```

```python
add_obj.add(2.3,3.2)
```

```python
add_obj.add("2","3")
```

```python
add_obj.add("2",3)
```
