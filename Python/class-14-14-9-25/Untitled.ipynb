## OOP (Object Oriented programming)

- Inheritance
- Abstraction (No implementation only define structure)
- Polymorphism (Class behavior more than one)
- Encapsulation (Hiding the information)


```python
class Parent():

    def __init__(self, eye_color, hair_color): # constructor
        self.eye_color = eye_color
        self.hair_color = hair_color
        

    def can_paint(self): # method
        print(f"have ability to draw/paint arts")
```

```python
par_obj = Parent("brown", "brown")
print(par_obj.eye_color)
print(par_obj.hair_color)
par_obj.can_paint()
```

```python
class Child(Parent):
    pass
```

```python
child_obj = Child("brown", "black")
print(child_obj.eye_color)
print(child_obj.hair_color)
child_obj.can_paint()
```

```python
print(child_obj.hair_color)
print(par_obj.hair_color)
```

```python
class Child(Parent):
    def __init__(self, eye_color, hair_color, skin_color):
        super().__init__(eye_color, hair_color)
        self.skin_color = skin_color

    def can_drive(self):
        print("Child can drive")
```

```python
child_obj = Child("brown", "black", "white")
print(child_obj.eye_color)
print(child_obj.hair_color)
child_obj.can_paint()
child_obj.can_drive()
```

```python
class Father():
    def __init__(self, height, cast):
        self.height = height
        self.cast = cast

    def anger(self):
        print(f"Father has anger")

    def hard_worker(self):
        print(f"Father is hard working")
```

```python
parent = Father("6.7", "baloch")

print(parent.height)
print(parent.cast)
```

```python
class Mother():
    def __init__(self, smile):
        self.smile = smile
```

```python
class Child(Father, Mother):
    def __init__(self, height, cast, smile, weight, body): # constructor overriding
        Father.__init__(self, height, cast)
        Mother.__init__(self, smile)
        self.weight = weight
        self.body = body


    def can_play_cricket(self):
        print(f"He can play good cricket")

    def can_cook(self):
        print(f"He can make good dishies")
```

```python
my_child = Child(cast="Baloch", height="6.7", smile="hehehe", weight="70kg", body="Body Builder")

print(my_child.height)
print(my_child.cast)
print(my_child.smile)
print(my_child.weight)
print(my_child.body)
my_child.anger()
my_child.can_cook()
my_child.can_play_cricket()
my_child.hard_worker()
```

## Overriding

```python
class Animals():

    def eat(self):
        print("Animals can eat") 
```

```python
ani_obj = Animals()
ani_obj.eat()
```

```python
class Birds(Animals):

    def eat(self): # method overriding
        print("Birds can eat")
```

```python
birds_obj = Birds()
birds_obj.eat()
```

## Overloading

```python
from typing import overload

class Adder():

    @overload
    def add(self, x:int, y:int) -> int:
        ...
    
    @overload
    def add(self, x:float, y:float) -> float:
        ...

    @overload
    def add(self, x:str, y:str) -> str:
        ...

    # Actual Function
    def add(self, x, y):
        if isinstance(x, int) and isinstance(y, int):
            return x+y
        elif isinstance(x, float) and isinstance(y, float):
            return x+y
        elif isinstance(x, str) and isinstance(y, str):
            return x+y
        else:
            print("Type Error")
```

```python
adder_obj = Adder()
adder_obj.add("2","3") # string case
```

```python
adder_obj.add(2,3.4) # type error
```

```python
adder_obj.add(2,3) # int case
```

```python
adder_obj.add(2.2,3.4) # float case
```
